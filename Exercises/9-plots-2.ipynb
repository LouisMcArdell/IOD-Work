```python
# Scatter plot

import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)  # Set a random seed for reproducibility
x = np.random.rand(50)  # Random x-values
y = np.random.rand(50)  # Random y-values

# Create a scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
plt.scatter(x, y, c='blue', marker='o', label='Random Data')

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Add a legend (optional)
plt.legend()

# Show the plot
plt.grid(True)  # Add a grid (optional)
plt.show()
```

```python
# Bar plot

import matplotlib.pyplot as plt

# Sample data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [1500, 2200, 1800, 1200]

# Create a bar plot
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
plt.bar(products, sales, color='blue')

# Add labels and a title
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales Bar Plot')

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels (optional)
plt.tight_layout()  # Adjust layout (optional)
plt.show()
```

```python
# Histogram

import matplotlib.pyplot as plt
import numpy as np

# Generate random data (in this case, a normal distribution)
np.random.seed(42)
data = np.random.normal(0, 1, 1000)

# Create a histogram
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
plt.hist(data, bins=20, color='blue', edgecolor='black')

# Add labels and a title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Show the plot
plt.grid(True)  # Add a grid (optional)
plt.tight_layout()  # Adjust layout (optional)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.rand(50)  # Random x-values
y = 2 * x + 1 + np.random.randn(50)  # Linear relationship with some noise

# Create a regplot
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
sns.regplot(x=x, y=y, color='blue', scatter_kws={'s': 50}, line_kws={'color': 'red'})

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Regression Line')

# Show the plot
plt.grid(True)  # Add a grid (optional)
plt.tight_layout()  # Adjust layout (optional)
plt.show()
```

```python
# Line plot

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 15]

# Create a line plot
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
plt.plot(x, y, marker='o', linestyle='-', color='blue', label='Data')

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')

# Add a legend (optional)
plt.legend()

# Show the plot
plt.grid(True)  # Add a grid (optional)
plt.tight_layout()  # Adjust layout (optional)
plt.show()
```

```python
# Box plot

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
data = [np.random.normal(0, 1, 100) for _ in range(4)]  # Create four datasets

# Create a box plot
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
sns.boxplot(data=data, palette="Set3")

# Add labels and a title
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.grid(True)  # Add a grid (optional)
plt.tight_layout()  # Adjust layout (optional)
plt.show()
```

```python
# pie chart

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [30, 15, 45, 10]  # Percentages or counts

# Create a pie chart
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
plt.pie(sizes, labels=categories, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'])

# Add a title
plt.title('Pie Chart Example')

# Show the plot
plt.tight_layout()  # Adjust layout (optional)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Heatmap plot

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for a correlation matrix
np.random.seed(42)
data = np.random.rand(5, 5)  # Create a 5x5 random matrix

# Create a heatmap
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
sns.heatmap(data, annot=True, cmap="coolwarm", fmt=".2f", square=True)

# Add a title
plt.title('Heatmap Example')

# Show the plot
plt.tight_layout()  # Adjust layout (optional)
plt.show()
```

```python
# Area plot

import matplotlib.pyplot as plt
import numpy as np

# Sample data
years = [2010, 2011, 2012, 2013, 2014, 2015]
values = [12, 25, 30, 22, 18, 28]

# Create an area plot
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
plt.fill_between(years, values, alpha=0.5, color='skyblue')

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Area Plot Example')

# Show the plot
plt.grid(True)  # Add a grid (optional)
plt.tight_layout()  # Adjust layout (optional)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random 3D data
np.random.seed(42)
n_points = 100
x = np.random.rand(n_points)
y = np.random.rand(n_points)
z = np.random.rand(n_points)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))  # Set the figure size (optional)
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='blue', marker='o')

# Add labels and a title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('3D Scatter Plot Example')

# Show the plot
plt.tight_layout()  # Adjust layout (optional)
plt.show()
```
