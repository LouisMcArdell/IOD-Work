## 1. Linear Regression

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
####################################################
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Importing the dataset
dataset = pd.read_csv('salary.csv',header=0)
```

```python
dataset
```

```python
X = dataset.iloc[:, 0].values
Y = dataset.iloc[:, -1].values
print(X)
```

```python
# Plot something of the above data
slope, intercept = np.polyfit(X, Y, 1)
y_fit = slope * X + intercept

plt.plot(X, y_fit, color='red')
plt.scatter(X,Y)
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.grid()
```

```python
#################################### The code below will be for later ##########################################################
```

```python
# Splitting the dataset into the Training set and Test set

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
regressor = LinearRegression()

# traiing data
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Visualising the Training set results

plt.figure(figsize=(8, 8))

# Train
plt.scatter(X_train, y_train, color = 'red', marker='v')

# Test (predicted) 
plt.scatter(X_test, y_test, color = 'green')

plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## 3. Multiple Linear regression example using scikit-learn

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# Load the Boston House prices dataset
boston = datasets.load_boston()
X = boston.data
y = boston.target
```

```python
print(boston.keys())

# data: contains the information for various houses
# target: prices of the house
# feature_names: names of the features
# DESCR: describes the dataset
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.5,random_state = 0)

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(Xtrain, ytrain)

# Make predictions using the testing set
ypred = regr.predict(Xtest)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)

#Intercept
print('Intercept: \n', regr.intercept_ )

# The mean absolute error
print("MAE = %5.3f" % mean_absolute_error(ytest, ypred))

# Explained variance score: 1 is perfect prediction
print("R^2 = %0.5f" % r2_score(ytest, ypred))

# The mean squared error
print("MSE = %5.3f" % mean_squared_error(ytest, ypred))
```

```python
# Plot outputs
fig, ax = plt.subplots()
ax.scatter(ytest, ypred, edgecolors = (0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw = 4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
