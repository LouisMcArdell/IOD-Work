## OLS Ordinary Least Squares Regression

```python
# 

from statsmodels.formula.api import ols
```

```python
# To simplify, y (endogenous) is the value you are trying to predict, while x (exogenous) represents the features you are 
# using to make the prediction.
```

### Simple Regression

```python
model = ols("sepal_length ~ petal_length", ir).fit()
```

```python
print(model.summary())
```

```python
# There is no value for AIC that can be considered good or bad because we simply use AIC as a way to compare regression 
# models. The model with the lowest AIC offers the best fit when comparing models.

# Akaike Information Criterion: AIC
```

```python
model2 = ols("sepal_length ~ sepal_width", ir).fit()
```

```python
print(model2.summary())
```

### Multiple Regression

```python
model3 = ols('sepal_width ~ petal_length + petal_length', ir).fit()
print(model3.summary())
```
